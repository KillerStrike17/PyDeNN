```python
RESNET_CONFIG = {_resnet: dict(input_layer='conv1', conv_layer='layer4', fc_layer='linear')
                 for _resnet in resnet.__dict__.keys()}
VGG_CONFIG = {_vgg: dict(input_layer='features', conv_layer='features')
              for _vgg in vgg.__dict__.keys()}
```

```python
MODEL_CONFIG = {**RESNET_CONFIG,**VGG_CONFIG}
```

```python
MODEL_CONFIG
```

```python
conv_layer = MODEL_CONFIG['ResNet18']['conv_layer']
input_layer = MODEL_CONFIG['ResNet18']['input_layer']
fc_layer = MODEL_CONFIG['ResNet18']['fc_layer']
```

```python
cam_extractors = [CAM(model, conv_layer, fc_layer), GradCAM(model, conv_layer),
                      GradCAMpp(model, conv_layer), SmoothGradCAMpp(model, conv_layer, input_layer),
                      ScoreCAM(model, conv_layer, input_layer), SSCAM(model, conv_layer, input_layer)]
```

```python
pil_img = Image.open(r'D:\Github\PyDeNN\DeNN\Tutorial\CAMs Visualization\car.jpg', mode='r').convert('RGB')
img_tensor = normalize(to_tensor(resize(pil_img, (32, 32))),[0.4914, 0.4822, 0.4465], [0.2023, 0.1994, 0.2010]).to(device=device)
```

```python
img_tensor.shape
```

```python
img_tensor.unsqueeze(0).shape
```

```python
fig, axes = plt.subplots(1, len(cam_extractors), figsize=(7, 2))
for idx, extractor in enumerate(cam_extractors):
    model.zero_grad()
    scores = model(img_tensor.unsqueeze(0))

    # Select the class index
    class_idx = scores.squeeze(0).argmax().item()

    # Use the hooked data to compute activation map
    activation_map = extractor(class_idx, scores).cpu()
    # Clean data
    extractor.clear_hooks()
    # Convert it to PIL image
    # The indexing below means first image in batch
    heatmap = to_pil_image(activation_map, mode='F')
    # Plot the result
    result = overlay_mask(pil_img, heatmap)

    axes[idx].imshow(result)
    axes[idx].axis('off')
    axes[idx].set_title(extractor.__class__.__name__, size=8)

plt.tight_layout()
plt.show()
```
